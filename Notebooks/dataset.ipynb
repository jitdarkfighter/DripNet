```python
from datasets import load_dataset

# streaming=True so we don't pull the whole 3.4 GB dataset up front.
dataset = load_dataset(
    "detection-datasets/fashionpedia",
    split="train",
    streaming=True,  # bool, not the string "true"
)
```

```python
import random
SEED = 50
random.seed(SEED)
```

```python
# Reservoir sampling
k = 1500
reservoir = []
import PIL
for i, sample in enumerate(dataset):
    if i < k:
        reservoir.append(sample)
    else:
        j = random.randint(0, i)
        if j < k:
            reservoir[j] = sample
```
