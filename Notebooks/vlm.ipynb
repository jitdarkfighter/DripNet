# Offline indexing pipeline

Caption + embed every image in `data/sampled_images/`, then build the two FAISS
indexes (caption-text + SigLIP image) the retriever fuses over. All logic lives in
`fashion_retrieval.indexer.index`; this notebook just drives it.

- Captioner: Qwen3-VL-2B (4-bit) -> structured JSON per image
- Image embedder: SigLIP image tower -> one `.npy` per image
- Text index: SigLIP text tower over `attrs_to_text(metadata)`
- Both stages are resumable (already-processed images are skipped).
